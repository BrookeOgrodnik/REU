```python
import sympy as sp
from sympy import sqrt
from sympy import log
from sympy import exp
from sympy import Matrix
from sympy import sin, cos
import matplotlib.pyplot as plt
from sympy import pi
```

```python
e = exp(1)
```

```python
def act(M, z):
    a,b,c,d = M[0,0], M[0,1], M[1,0], M[1,1]
    x,y = z
    q = c*(x**2+y**2)+d**2+2*c*d*x
    r = a*c*(x**2+y**2)+b*d+(a*d+b*c)*x
    return (r/q,y/q)
```

```python
def getCirc(z1,z2):
    x1,y1 = z1
    x2,y2 = z2
    if x1==x2:
        return ('inf', x1)
    x0 = (y2**2+x2**2-y1**2-x1**2)/(2*(x2-x1))
    r = sqrt((x1-x0)**2+y1**2)
    return (r,x0)
```

```python
def getDist(z1,z2):
    x1,y1 = z1
    x2,y2 = z2
    r,x0 = getCirc(z1,z2)
    if r=='inf':
        dist = log(y2/y1)
    else:
        dist = log(sqrt((r+x2-x0)*(r+x0-x1)/(r+x1-x0)/(r+x0-x2)))
    return dist
```

```python
def getDistortion(M,z1,z2):
    l = getDist(z1,z2)
    l1 = getDist(act(M,z1),act(M,z2))
    l = abs(l.evalf())
    l1 = abs(l1.evalf())
    return l1/l
```

```python
def getRot(angle):
    return Matrix([[cos(angle).evalf(), -sin(angle).evalf()],[sin(angle).evalf(), cos(angle).evalf()]])
```

```python
def getFlow(t):
    a,b = e**(t/2), e**(-t/2)
    a = a.evalf()
    b = b.evalf()
    return Matrix([[a,0],[0,b]])
```

```python
N = 5000
x_values,y_values = [],[]
for i in range(N+1):
    x_values.append(i/N)
    y_values.append(getDistortion(getRot(2*i*pi/N), (0,1), (0,e**5)))
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
plt.plot(x_values,y_values)
```

```python
Q = 15
for i in range(Q+1):
    N = 1000
    x_values,y_values = [],[]
    t = 5
    theta = i/Q+pi/13
    rot = getRot(2*theta*pi)
    step = t/N
    for i in range(-N, N+1):
        t = i*step
        M = getFlow(t)
        z = act(M, (0,1))
        x,y = act(rot, z)
        x_values.append(x)
        y_values.append(y)
    fig, ax = plt.subplots(figsize=(20, 10))
    plt.scatter(x_values,y_values)
    plt.plot([0,0.00001],[1,1.00001],color = 'red', markersize = 50)
    plt.show()
```
